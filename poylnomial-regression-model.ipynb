```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("Position_Salaries.csv") 
data 
```

## Splitting the Data 

```python
x = data.iloc[:, 1: -1].values
y = data.iloc[: , -1].values
```

```python
x 
```

```python
y
```

## Linear Regression Model 

```python
from sklearn.linear_model import LinearRegression 

lin_reg = LinearRegression() 
lin_reg.fit(x,y) 
```

## Visualise the LR Model 

```python
plt.scatter(x,y, color = "red") 
plt.plot(x, lin_reg.predict(x), color = "blue") 
plt.xlabel("Input")
plt.ylabel("Predictions")
plt.show()
```

## Polynomial Linear Regression 

```python
from sklearn.preprocessing import PolynomialFeatures 
poly_reg = PolynomialFeatures(degree = 2 ) 
x_poly = poly_reg.fit_transform(x)
```

```python
x_poly 
```

```python
poly_linear_reg = LinearRegression()
```

```python
poly_linear_reg.fit(x_poly, y) 
```

## Visualizing Polynomial Liner Regression 

```python
plt.scatter(x,y, color = "red") 
plt.plot(x, poly_linear_reg.predict(poly_reg.fit_transform(x)), color = "blue") 
plt.xlabel("Input")
plt.ylabel("Predictions")
plt.show()
```

# Effect of Polynomial Degree in Polynomial Regression

In polynomial regression, the degree of the polynomial features determines the complexity of the model. Increasing the degree can lead to a better fit to the training data, but it may also result in overfitting, where the model learns noise in the training data rather than the underlying pattern. Conversely, decreasing the degree may result in underfitting, where the model is too simple to capture the underlying pattern.

## Increasing Degree

When we increase the degree of polynomial features:

- The model becomes more flexible and can capture more complex patterns in the data.
- The training error tends to decrease, as the model can better fit the training data.
- However, there is a risk of overfitting, where the model learns noise in the training data and performs poorly on unseen data.

## Decreasing Degree

When we decrease the degree of polynomial features:

- The model becomes less flexible and may not capture complex patterns in the data.
- The training error may increase, as the model may underfit the training data.
- However, there is a lower risk of overfitting, and the model may generalize better to unseen data.

Therefore, it's important to choose an appropriate degree of polynomial features based on the trade-off between model complexity and generalization performance.
